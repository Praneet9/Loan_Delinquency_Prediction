```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_data = pd.read_csv('train.csv')
train_data.head()
```

```python
train_data.describe()
```

```python
train_data.columns
```

```python
train_data['loan_purpose'].unique()
```

```python
target = train_data['m13']
cleaned_train = train_data.drop(labels=['loan_id','source','financial_institution','origination_date','first_payment_date','m13'], axis=1)
cleaned_train = pd.concat([cleaned_train, pd.get_dummies(cleaned_train['loan_purpose'], prefix='loan_purpose')], axis=1)
cleaned_train = cleaned_train.drop(labels=['loan_purpose'], axis=1)
cleaned_train.head()
```

```python
cleaned_train.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()
```

```python
X = np.asarray(cleaned_train)
y = np.asarray(target)
```

```python
classifier.fit(X, y)
```

```python
test_data = pd.read_csv('test.csv')
test_data.head()
```

```python
cleaned_test = test_data.drop(labels=['loan_id','source','financial_institution','origination_date','first_payment_date'], axis=1)
cleaned_test = pd.concat([cleaned_test, pd.get_dummies(cleaned_test['loan_purpose'], prefix='loan_purpose')], axis=1)
cleaned_test = cleaned_test.drop(labels=['loan_purpose'], axis=1)
cleaned_test.head()
```

```python
X_test = np.asarray(cleaned_test)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
indices = np.array([idx for idx in range(1,len(y_pred)+1)])
```

```python
submission_data = pd.DataFrame({'loan_id': indices, 'm13': y_pred}).set_index('loan_id')
submission_data.head()
```

```python
submission_data.to_csv('basic_submission.csv')
```
